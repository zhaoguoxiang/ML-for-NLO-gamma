```python
def xyz2mols(xyzfiles):
    mols = []
    for xyzfile in xyzfiles:
        bel_mol = pybel.readfile('xyz',str(xyzfile)).__next__()
        smi = smi = bel_mol.write('smi').split('\t')[0]
        mol = Chem.MolFromSmiles(smi,sanitize=False)
        mols.append(mol)
    
    return mols
```

```python
xyzfiles = list(gammalogs.rglob('*.xyz'))

mols = xyz2mols(xyzfiles)
```

```python
mols
```

```python
from rdkit.Chem import MACCSkeys
```

```python
fingerprints = [MACCSkeys.GenMACCSKeys(molecule) for molecule in mols]
ls = [list(fp.ToBitString()) for fp in fingerprints]
X = np.array(ls)
```
